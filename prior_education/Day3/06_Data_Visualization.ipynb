```python
import matplotlib.pyplot as plt
import numpy as np

# 데이터 준비
x = np.linspace(0, 10, 100)
y = np.sin(x)
data = np.random.randn(1000)
labels = ['A', 'B', 'C']
values = [15, 30, 20]
```

```python
# 2행 2열의 서브플롯 생성
fig, axes = plt.subplots(2, 2, figsize=(10,8))
```

```python
# [0, 0] 선 그래프
axes[0, 0].plot(x, y, color='blue')
axes[0, 0].set_title("① 선 그래프")

# [0, 1] 산점도
axes[0, 1].scatter(x, y, color='green')
axes[0, 1].set_title("② 산점도")

# [1, 0] 히스토그램
axes[1, 0].hist(data, bins=30, color='orange', edgecolor='black')
axes[1, 0].set_title("③ 히스토그램")

# [1, 1] 막대 그래프
axes[1, 1].bar(labels, values, color='purple')
axes[1, 1].set_title("④ 막대 그래프")

# 전체 제목과 레이아웃 조정
fig.suptitle("matplotlib의 다중 그래프(subplots)", fontsize=16)
plt.tight_layout(rect=[0, 0, 1, 0.95])  # 전체 제목과 겹치지 않도록 조정
plt.show()
```

```python
# tips 데이터셋을 선정
tips = sns.load_dataset('tips')
```

```python
# 산점도 분석 출력
# 하지만 해석하는게 제일 중요하다!!
sns.scatterplot(data=tips, x='total_bill', y='tip')
plt.show()
```

```python
# 선 그래프
sns.lineplot(data=tips, x='total_bill', y='tip')
```

### `sns.lineplot` 분석 인사이트 요약

#### - 평균 팁의 증가 추세
총 계산 금액이 커질수록 평균적으로 지불되는 팁도 꾸준히 증가하는 경향을 보입니다.

#### - 고액 계산에서의 불확실성 증가
계산 금액이 매우 큰 구간에서는 데이터가 부족하여 평균 팁에 대한 통계적 신뢰도가 낮아집니다.


```python
# histogram 표현
sns.histplot(data=tips, x='total_bill', bins=10)
```

### `sns.histplot` 분석 인사이트 요약

#### - 분포 형태: 오른쪽으로 긴 꼬리
전체 계산 금액(`total_bill`)의 분포는 대부분 낮은 금액대에 몰려 있고, 고액 결제는 드물게 나타나는 **오른쪽으로 치우친(right-skewed)** 형태를 보입니다.

#### - 핵심 구간: $10 ~ $20
가장 빈번하게 발생하는 계산 금액 구간은 **$10에서 $20 사이**로, 이곳에 대부분의 고객이 집중되어 있음을 알 수 있습니다.


```python
# KDE 그래프 포현
# fill=True : 그래프 채우기
sns.kdeplot(data=tips, x='total_bill', fill=True)
```

### `sns.kdeplot` 분석 인사이트 요약

#### - 가장 가능성 높은 금액대
곡선의 **가장 높은 봉우리(peak)**는 약 **$15 ~ $20** 사이에 위치하며, 이 금액대의 계산이 가장 빈번하게 발생함을 의미합니다.

#### - 꼬리 모양 (Tail Shape)
그래프가 **오른쪽으로 긴 꼬리**를 가지는 것은, 대부분의 계산은 낮은 금액대에 집중되어 있지만, 간혹 매우 높은 금액의 계산도 발생한다는 것을 보여줍니다.


```python
# 박스플롯 표현
sns.boxplot(data=tips, x='day', y='tip')
```

### `sns.boxplot` 분석 인사이트 요약

#### - 주말 팁이 더 높다 (Median Comparison)
**토요일(Sat)과 일요일(Sun)**의 중앙값(상자 안의 선)이 목요일(Thur)과 금요일(Fri)보다 높아, 주말에 **평균적으로 더 높은 팁**을 주는 경향을 보입니다.

#### - 특별히 높은 팁 (Outliers)
모든 요일에 평균을 훨씬 웃도는 **이상치(outlier) 팁**이 존재하며, 특히 주말에 더 높고 많은 이상치가 관찰됩니다. 이는 서비스에 매우 만족한 소수의 고객이 존재함을 시사합니다.


```python
# 바이올린 플롯 표현
sns.violinplot(data=tips, x='day', y='tip')
```

### `sns.violinplot` 분석 인사이트 요약

#### 분포의 밀도를 한눈에 파악
바이올린의 **"넓은" 부분**은 해당 값에 데이터가 **많이 몰려 있음**을 의미합니다. 모든 요일에서 팁은 약 $2 ~ $3 사이에 가장 많이 분포하는 것을 시각적으로 확인할 수 있습니다.

#### 팁의 다양성
**주말(Sat, Sun)**의 바이올린이 평일(Thur, Fri)보다 **더 길고 넓게 퍼져** 있습니다. 이는 주말에 팁이 평균적으로 높을 뿐만 아니라, 매우 적은 팁부터 많은 팁까지 **더 다양하게 분포**하고 있음을 의미합니다.


```python
# 막대그래프 표현
sns.barplot(data=tips, x='day', y='tip')
```

### `sns.barplot` 분석 인사이트 요약

#### - 요일별 평균 팁 비교
각 막대의 높이는 해당 요일의 **평균 팁**을 나타냅니다. **주말(일요일, 토요일)**의 평균 팁이 평일(목요일, 금요일)보다 높다는 것을 한눈에 파악할 수 있습니다.

#### - 평균값의 신뢰도
각 막대 위의 검은 선(오차 막대, error bar)은 **평균값의 신뢰 구간**을 의미합니다. 이 선이 길수록 해당 요일의 평균값에 대한 통계적 불확실성이 크다는 것을 나타냅니다.


```python
# 카운트플롯 표현
sns.countplot(data=tips, x='day')
```

### `sns.countplot` 분석 인사이트 요약

#### - 가장 바쁜 날: 토요일
**토요일(Sat)** 막대가 가장 높아, 레스토랑에 손님이 가장 많이 방문했음을 의미합니다.

#### - 가장 한가한 날: 금요일
**금요일(Fri)** 막대가 가장 낮아, 데이터상으로는 방문객 수가 가장 적은 요일임을 알 수 있습니다.


```python
# 조인트플롯 표현
sns.jointplot(data=tips, x='total_bill', y='tip', kind='scatter')
```

### `sns.jointplot` 분석 인사이트 요약

#### - 두 변수 관계와 분포를 동시에 분석
이 플롯은 두 변수(`total_bill`, `tip`)의 관계를 보여주는 **산점도(가운데)**와 각 변수의 **개별 분포(위쪽, 오른쪽 히스토그램)**를 한 번에 결합하여 보여줍니다.

#### - 중심부: 양의 상관관계
가운데 산점도는 `total_bill`과 `tip` 사이에 강한 **양의 상관관계**가 있음을 다시 한번 확인시켜 줍니다.

#### - 주변부: 각 변수의 분포
**위쪽 히스토그램**은 `total_bill`이 $10~$20에 집중된 분포를, **오른쪽 히스토그램**은 `tip`이 $2~$3에 집중된 분포를 보여줍니다.


```python
# Pairplot 표현
sns.pairplot(tips, hue='day')
```

### `sns.pairplot` 분석 인사이트 요약

#### - 데이터셋 전체 관계 조망
이 그래프는 데이터셋의 모든 **숫자형 변수들 간의 관계를 한눈에 파악**할 수 있는 그리드(grid)를 생성합니다.

#### - 대각선: 개별 변수의 분포
대각선 위치에는 각 변수(`total_bill`, `tip`, `size`)의 **개별 분포**를 보여주는 히스토그램이 그려집니다.

#### - 비대각선: 변수 쌍의 관계와 `hue`를 통한 비교
- **비대각선** 위치에는 변수 쌍 간의 관계를 보여주는 **산점도**가 그려집니다.
- `hue='day'` 옵션은 각 점을 **요일별로 다른 색상으로 구분**하여, 요일에 따라 변수 간의 관계나 분포가 어떻게 달라지는지 비교할 수 있게 해줍니다.
