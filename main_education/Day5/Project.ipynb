```python
# YOLO 모델 로드
model = YOLO("yolo11n.pt")

# 차량 / 사람 클래스 ID (COCO 기준)
vehicle_classes = [2, 3, 5, 7]  # car, motorcycle, bus, truck
person_class = 0

video_path = "video.mp4"

all_rois = [[(353, 621), (325, 664), (1034, 673), (1007, 630), (354, 621)], [(1076, 682), (1114, 718), (1275, 714), (1276, 686), (1080, 680)], [(1, 656), (285, 662), (256, 717), (4, 718), (3, 655)]]
```

```python
# 필요한 라이브러리 추가
from shapely.geometry import Polygon
import cv2
import numpy as np
# from ultralytics import YOLO # YOLO 모델 로드를 위해 필요

# 2. 객체 추적 및 위험 감지 단계
print("\n객체 추적을 시작합니다...")
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
    exit()

# 위험 근접도 임계값 설정
# 사람과 차량이 이 픽셀 거리보다 가까우면 위험으로 간주 (영상에 맞게 튜닝 필요)
DANGER_PROXIMITY_THRESHOLD = 150

# 차량 하단 감지를 위한 비율 설정
# 차량 바운딩 박스의 하단 30%만 교차 검사에 사용
VEHICLE_BOTTOM_RATIO = 0.3

# ROI를 미리 Shapely Polygon 객체로 변환 (효율성 향상)
try:
    shapely_rois = [Polygon(roi) for roi in all_rois]
except Exception as e:
    print(f"Shapely Polygon 생성 오류: {e}")
    print("ROI 좌표가 3개 이상이고 닫힌 형태인지 확인하세요.")
    exit()

# 처리된 위험 '쌍'을 기록하여 중복 카운트 방지
total_danger_events = 0
processed_danger_pairs = set()

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("비디오 스트림이 종료되었습니다.")
        break

    frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES))
    print(f"\n--- 프레임 #{frame_number} 처리 시작 ---")

    roi_occupancy = [{'persons': {}, 'vehicles': {}} for _ in all_rois]
    results = model.track(frame, persist=True, conf=0.4, iou=0.5, classes=[person_class] + vehicle_classes)

    num_detected = len(results[0].boxes) if results[0].boxes is not None and results[0].boxes.id is not None else 0
    print(f"[탐지] 이 프레임에서 {num_detected}개의 추적 가능한 객체 탐지됨")

    if results[0].boxes is not None and results[0].boxes.id is not None:
        for box in results[0].boxes:
            cls_id = int(box.cls[0])
            track_id = int(box.id[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            
            # 1. 객체의 바운딩 박스를 Shapely Polygon으로 생성
            # 사람인 경우, 전체 바운딩 박스를 교차 검사에 사용
            if cls_id == person_class:
                intersection_check_polygon = Polygon([(x1, y1), (x2, y1), (x2, y2), (x1, y2)])
            # 차량인 경우, 바운딩 박스의 하단 부분만 교차 검사에 사용
            elif cls_id in vehicle_classes:
                height = y2 - y1
                y1_bottom_part = y2 - int(height * VEHICLE_BOTTOM_RATIO)
                intersection_check_polygon = Polygon([(x1, y1_bottom_part), (x2, y1_bottom_part), (x2, y2), (x1, y2)])
            else:
                # 사람이나 차량이 아니면 건너뜀
                continue

            for i, roi_polygon in enumerate(shapely_rois):
                # 2. 두 Polygon이 교차하는지 확인
                if roi_polygon.intersects(intersection_check_polygon):
                    print(f"  [교차!] 객체 ID {track_id} ({model.names[cls_id]})가 ROI #{i+1}와 교차함")
                    
                    obj_data = {'box': (x1, y1, x2, y2), 'cls': cls_id}
                    if cls_id == person_class:
                        roi_occupancy[i]['persons'][track_id] = obj_data
                    elif cls_id in vehicle_classes:
                        roi_occupancy[i]['vehicles'][track_id] = obj_data
                    break

    print("[상태] 현재 프레임 ROI 점유 상태:")
    for i, occupancy in enumerate(roi_occupancy):
        person_ids = list(occupancy['persons'].keys())
        vehicle_ids = list(occupancy['vehicles'].keys())
        if person_ids or vehicle_ids:
            print(f"  - ROI #{i+1}: 사람들(ID:{person_ids}), 차량들(ID:{vehicle_ids})")

    # 위험 상황 평가
    dangerous_object_ids = set()
    is_any_roi_dangerous = [False] * len(all_rois)

    for i, occupancy in enumerate(roi_occupancy):
        persons_in_roi = occupancy['persons']
        vehicles_in_roi = occupancy['vehicles']

        # 1. ROI에 사람과 차량이 모두 존재하는지 먼저 확인
        if persons_in_roi and vehicles_in_roi:
            is_any_roi_dangerous[i] = True
            print(f"[!!] 잠재적 위험! ROI #{i+1}에 사람과 차량 동시 존재! 근접도 검사 시작...")

            # 2. 모든 사람-차량 쌍에 대해 '근접도'를 검사
            for p_id, p_data in persons_in_roi.items():
                for v_id, v_data in vehicles_in_roi.items():
                    # 각 객체의 중심점 계산
                    p_box = p_data['box']
                    v_box = v_data['box']
                    p_center = np.array([(p_box[0] + p_box[2]) / 2, (p_box[1] + p_box[3]) / 2])
                    v_center = np.array([(v_box[0] + v_box[2]) / 2, (v_box[1] + v_box[3]) / 2])
                    
                    # 두 객체 중심점 간의 거리 계산
                    distance = np.linalg.norm(p_center - v_center)

                    # 3. 거리가 임계값보다 가까운 경우에만 '실제 위험'으로 판단
                    if distance < DANGER_PROXIMITY_THRESHOLD:
                        is_any_roi_dangerous[i] = True # ROI를 위험 상태로 표시
                        danger_pair = (p_id, v_id)

                        # 4. 이전에 처리되지 않은 새로운 위험 쌍인 경우에만 카운트 증가
                        if danger_pair not in processed_danger_pairs:
                            total_danger_events += 1
                            processed_danger_pairs.add(danger_pair)
                            print(f"    [+] 실제 위험 감지! (사람 ID:{p_id}, 차량 ID:{v_id}, 거리:{distance:.1f}px). 총 위험 건수: {total_danger_events}")
                        
                        # 5. 시각화를 위해 위험에 관련된 객체 ID를 추가
                        dangerous_object_ids.add(p_id)
                        dangerous_object_ids.add(v_id)

    # 화면 상단에 누적된 위험 건수 표시
    cv2.putText(frame, f"DANGER COUNT: {total_danger_events}", (50, 50), cv2.FONT_HERSHEY_TRIPLEX, 1.2, (0, 0, 255), 2)

    # 시각화
    if results[0].boxes is not None and results[0].boxes.id is not None:
        for box in results[0].boxes:
            track_id = int(box.id[0])
            cls_id = int(box.cls[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            
            if track_id in dangerous_object_ids:
                color = COLOR_DANGER_OBJ
            else:
                color = COLOR_PERSON if cls_id == person_class else COLOR_VEHICLE
            
            label = f"ID {track_id} {model.names[cls_id]}"
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    for i, roi in enumerate(all_rois):
        roi_color = COLOR_ROI_DANGER if is_any_roi_dangerous[i] else COLOR_ROI_NORMAL
        cv2.polylines(frame, [np.array(roi, np.int32)], isClosed=True, color=roi_color, thickness=3)
        if is_any_roi_dangerous[i]:
            (x, y, w, h) = cv2.boundingRect(np.array(roi, np.int32))
            cv2.putText(frame, f"DANGER on Crosswalk {i+1}!", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, COLOR_ROI_DANGER, 2)

    cv2.imshow("Crosswalk Safety Monitor", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# 필요한 라이브러리 추가
from shapely.geometry import Polygon
import cv2
import numpy as np
import time # 시간 추적을 위해 추가

# 객체 추적 및 위험 감지 단계
print("\n객체 추적을 시작합니다...")
cap = cv2.VideoCapture(video_path)
if not cap.isOpened():
    print(f"Error: Could not open video file {video_path}")
    exit()

# 비디오 저장을 위한 설정
frame_width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(cap.get(cv2.CAP_PROP_FPS))

# VideoWriter 객체 생성
output_path = 'output.mp4'
fourcc = cv2.VideoWriter_fourcc(*'mp4v') # 또는 'XVID'
out = cv2.VideoWriter(output_path, fourcc, fps, (frame_width, frame_height))


# 위험 근접도 임계값 설정
DANGER_PROXIMITY_THRESHOLD = 150

# 차량 하단 감지를 위한 비율 설정
VEHICLE_BOTTOM_RATIO = 0.3

# ROI를 미리 Shapely Polygon 객체로 변환
try:
    shapely_rois = [Polygon(roi) for roi in all_rois]
except Exception as e:
    print(f"Shapely Polygon 생성 오류: {e}")
    print("ROI 좌표가 3개 이상이고 닫힌 형태인지 확인하세요.")
    exit()

# 처리된 위험 '쌍' 및 마지막 위험 감지 시간 기록
total_danger_events = 0
processed_danger_pairs = set()
last_danger_time = 0  # 마지막 위험 감지 시간 초기화

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        print("비디오 스트림이 종료되었습니다.")
        break

    frame_number = int(cap.get(cv2.CAP_PROP_POS_FRAMES))
    print(f"\n--- 프레임 #{frame_number} 처리 시작 ---")

    roi_occupancy = [{'persons': {}, 'vehicles': {}} for _ in all_rois]
    results = model.track(frame, persist=True, conf=0.4, iou=0.5, classes=[person_class] + vehicle_classes)

    num_detected = len(results[0].boxes) if results[0].boxes is not None and results[0].boxes.id is not None else 0
    print(f"[탐지] 이 프레임에서 {num_detected}개의 추적 가능한 객체 탐지됨")

    if results[0].boxes is not None and results[0].boxes.id is not None:
        for box in results[0].boxes:
            cls_id = int(box.cls[0])
            track_id = int(box.id[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            
            if cls_id == person_class:
                intersection_check_polygon = Polygon([(x1, y1), (x2, y1), (x2, y2), (x1, y2)])
            elif cls_id in vehicle_classes:
                height = y2 - y1
                y1_bottom_part = y2 - int(height * VEHICLE_BOTTOM_RATIO)
                intersection_check_polygon = Polygon([(x1, y1_bottom_part), (x2, y1_bottom_part), (x2, y2), (x1, y2)])
            else:
                continue

            for i, roi_polygon in enumerate(shapely_rois):
                if roi_polygon.intersects(intersection_check_polygon):
                    print(f"  [교차!] 객체 ID {track_id} ({model.names[cls_id]})가 ROI #{i+1}와 교차함")
                    obj_data = {'box': (x1, y1, x2, y2), 'cls': cls_id}
                    if cls_id == person_class:
                        roi_occupancy[i]['persons'][track_id] = obj_data
                    elif cls_id in vehicle_classes:
                        roi_occupancy[i]['vehicles'][track_id] = obj_data
                    break

    print("[상태] 현재 프레임 ROI 점유 상태:")
    for i, occupancy in enumerate(roi_occupancy):
        person_ids = list(occupancy['persons'].keys())
        vehicle_ids = list(occupancy['vehicles'].keys())
        if person_ids or vehicle_ids:
            print(f"  - ROI #{i+1}: 사람들(ID:{person_ids}), 차량들(ID:{vehicle_ids})")

    # 위험 상황 평가
    dangerous_object_ids = set()
    is_any_roi_dangerous = [False] * len(all_rois)

    for i, occupancy in enumerate(roi_occupancy):
        persons_in_roi = occupancy['persons']
        vehicles_in_roi = occupancy['vehicles']

        if persons_in_roi and vehicles_in_roi:
            print(f"[!!] 잠재적 위험! ROI #{i+1}에 사람과 차량 동시 존재! 근접도 검사 시작...")
            for p_id, p_data in persons_in_roi.items():
                for v_id, v_data in vehicles_in_roi.items():
                    p_box = p_data['box']
                    v_box = v_data['box']
                    p_center = np.array([(p_box[0] + p_box[2]) / 2, (p_box[1] + p_box[3]) / 2])
                    v_center = np.array([(v_box[0] + v_box[2]) / 2, (v_box[1] + v_box[3]) / 2])
                    distance = np.linalg.norm(p_center - v_center)

                    if distance < DANGER_PROXIMITY_THRESHOLD:
                        is_any_roi_dangerous[i] = True
                        danger_pair = (p_id, v_id)
                        if danger_pair not in processed_danger_pairs:
                            total_danger_events += 1
                            processed_danger_pairs.add(danger_pair)
                            last_danger_time = time.time()  # 새로운 위험 감지 시간 기록
                            print(f"    [+] 실제 위험 감지! (사람 ID:{p_id}, 차량 ID:{v_id}, 거리:{distance:.1f}px). 총 위험 건수: {total_danger_events}")
                        
                        dangerous_object_ids.add(p_id)
                        dangerous_object_ids.add(v_id)

    # 우측 상단 UI 표시
    # Danger Count 표시
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    thickness = 2
    text_color = (0, 0, 255)  # Red in BGR
    bg_color_count = (255, 255, 255) # White
    margin = 10
    frame_width = frame.shape[1]

    count_text = f"DANGER COUNT: {total_danger_events}"
    (text_w, text_h), baseline = cv2.getTextSize(count_text, font, font_scale, thickness)

    # 배경 사각형과 텍스트 위치 계산
    rect_x2 = frame_width - margin
    rect_y1 = margin
    rect_x1 = rect_x2 - text_w - margin
    rect_y2 = rect_y1 + text_h + baseline + margin
    text_x = rect_x1 + (margin // 2)
    text_y = rect_y1 + text_h + (margin // 2)

    # 배경과 텍스트 그리기
    cv2.rectangle(frame, (rect_x1, rect_y1), (rect_x2, rect_y2), bg_color_count, cv2.FILLED)
    cv2.putText(frame, count_text, (text_x, text_y), font, font_scale, text_color, thickness)

    # 2. 새로운 위험 감지 후 3초간 'Danger' 경고 표시
    if time.time() - last_danger_time < 2:
        danger_text = "Danger"
        bg_color_danger = (0, 0, 0) # Black
        (text_w_d, text_h_d), baseline_d = cv2.getTextSize(danger_text, font, font_scale, thickness)
        
        # Danger Count 상자 아래에 위치
        rect_d_y1 = rect_y2 + margin
        rect_d_x2 = frame_width - margin
        rect_d_x1 = rect_d_x2 - text_w_d - margin
        rect_d_y2 = rect_d_y1 + text_h_d + baseline_d + margin
        text_d_x = rect_d_x1 + (margin // 2)
        text_d_y = rect_d_y1 + text_h_d + (margin // 2)

        cv2.rectangle(frame, (rect_d_x1, rect_d_y1), (rect_d_x2, rect_d_y2), bg_color_danger, cv2.FILLED)
        cv2.putText(frame, danger_text, (text_d_x, text_d_y), font, font_scale, text_color, thickness)


    # 시각화
    if results[0].boxes is not None and results[0].boxes.id is not None:
        for box in results[0].boxes:
            track_id = int(box.id[0])
            cls_id = int(box.cls[0])
            x1, y1, x2, y2 = map(int, box.xyxy[0])
            
            if track_id in dangerous_object_ids:
                color = COLOR_DANGER_OBJ
            else:
                color = COLOR_PERSON if cls_id == person_class else COLOR_VEHICLE
            
            label = f"ID {track_id} {model.names[cls_id]}"
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    for i, roi in enumerate(all_rois):
        roi_color = COLOR_ROI_DANGER if is_any_roi_dangerous[i] else COLOR_ROI_NORMAL
        cv2.polylines(frame, [np.array(roi, np.int32)], isClosed=True, color=roi_color, thickness=3)
        if is_any_roi_dangerous[i]:
            (x, y, w, h) = cv2.boundingRect(np.array(roi, np.int32))
            cv2.putText(frame, f"DANGER on Crosswalk {i+1}!", (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, COLOR_ROI_DANGER, 2)

    # 처리된 프레임을 비디오 파일에 저장
    out.write(frame)
    
    cv2.imshow("Crosswalk Safety Monitor", frame)
    if cv2.waitKey(1) & 0xFF == 27:
        break

# 자원 해제
cap.release()
out.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```

```python
# YOLO 모델 로드
model = YOLO("yolo11n.pt")

# 차량 / 사람 클래스 ID (COCO 기준)
vehicle_classes = [2, 3, 5, 7]  # car, motorcycle, bus, truck
person_class = 0

video_path = "data/video.mp4"
cap = cv2.VideoCapture(video_path)

while cap.isOpened():
    ret, frame = cap.read()
    if not ret:
        break

    # YOLO 추적 모드 실행
    results = model.track(frame, persist=True, conf=0.5, iou=0.5)

    vehicles_on_crosswalk = []
    persons_on_crosswalk = []

    if results[0].boxes.id is not None:
        for box in results[0].boxes:
            cls_id = int(box.cls[0])
            track_id = int(box.id[0])  # 추적 ID
            x1, y1, x2, y2 = map(int, box.xyxy[0])

            # 바운딩 박스 중심 (사람은 발 위치, 차량은 하단 중심)
            if cls_id == person_class:
                cx, cy = int((x1 + x2) / 2), y2  # 보행자: 발 위치
            elif cls_id in vehicle_classes:
                cx, cy = int((x1 + x2) / 2), y2  # 차량: 하단 중심
            else:
                continue

            # ROI 내부 여부 판단 (여러 ROI 각각 검사)
            inside = False
            for roi in all_rois:
                if cv2.pointPolygonTest(np.array(roi, np.int32), (cx, cy), False) >= 0:
                    inside = True
                    break

            # ROI 내부일 경우 리스트에 저장
            if inside:
                if cls_id == person_class:
                    persons_on_crosswalk.append(track_id)
                else:
                    vehicles_on_crosswalk.append(track_id)

            # 바운딩 박스 + ID 시각화
            color = (0, 255, 0) if cls_id == person_class else (255, 0, 0)
            cv2.rectangle(frame, (x1, y1), (x2, y2), color, 2)
            cv2.putText(frame, f"ID {track_id}", (x1, y1 - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, color, 2)

    danger = False
    if len(persons_on_crosswalk) > 0 and len(vehicles_on_crosswalk) > 0:
        danger = True
        cv2.putText(frame, "⚠ DANGER: Person + Vehicle on crosswalk!",
                    (50, 100), cv2.FONT_HERSHEY_SIMPLEX, 1.0, (0, 0, 255), 3)

    # ROI 시각화 (여러 ROI 모두 그림)
    for roi in all_rois:
        cv2.polylines(frame, [np.array(roi, np.int32)], isClosed=True, color=(0, 255, 255), thickness=2)

    cv2.imshow("Crosswalk Monitor", frame)
    if cv2.waitKey(1) & 0xFF == 27:  # ESC 종료
        break

cap.release()
cv2.destroyAllWindows()
cv2.waitKey(1)
```
