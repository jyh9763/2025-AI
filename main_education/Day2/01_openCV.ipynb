#### Python기반 OpenCV를 이용한 영상(이미지) 처리: Computer Vision
OpenCV(Open Source Vision Library)
- 이미지 처리나 computer vision에서 사용되는 library
- 주로 Python기반에서 사용됨
- 다양한 프로그래밍 언어(C, C++, JAVA)를 지원함

[특징]
1. 이미지 읽기, 쓰기, 변환, 필터링과 같은 이미지 처리 가능
2. 비디오 파일을 웹캠을 통해서 비디오 스트리밍을 처리할 수 있음
    > 비디오 프레임을 개별적으로 처리하여 객체 탐지(Object Detection), 객체 추적(Object Tracking) 등이 가능
3. 다양한 객체 탐지(Object Detection) 알고리즘을 지원
    > Haar Cascade, YOLO(You Only Look Once),SSD(Single Shot Multibox Detector) 등을 사용하여
    얼굴, 사람, 자동차 등의 객체를 실시간으로 탐지할 수 있음
4. Tensorflow, pyTorch 등의 프레임워크와 통합하여 사용 가능
5. 예지 정비, 예지 검출, 윤곽선 검출, 모션 추척, 스테레오 비젼 등 다양한 컴퓨터 비젼 알고리즘 제공

[사용]
1. 출입 통제 시스템(얼굴 인식 등)
2. 보안 시스템
3. 자율 주행 - 차선, 보행자 등 인식
4. 제조 공장 - 불량 검출, 제품 품질 검사, 예지 정비 등
5. 의료 - 흉부X-ray, 암 등 병변을 탐지하거나 의료 진단 활용
6. 증강 현실

```python
# 라이브러리 불러오기
import cv2
import matplotlib.pyplot as plt

# 이미지 불러오기
image = cv2.imread('./data/image.jpg')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))   # opencv는 BGR로 이미지를 읽기 때문에 -> RGB로 변환
plt.axis('off')  # 축 제거
plt.show()
```

```python
# 이미지 불러오기
image = cv2.imread('./data/image.jpg')

### 이미지 색상 변환
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)  # 흑백으로 변환
plt.imshow(image, cmap='gray')  # 흑백 이미지 표시
plt.axis('off')  # 축 제거

# 변환된 이미지 저장
cv2.imwrite('image2.png', image)
plt.show()
```

```python
# 라이브러리 불러오기
import cv2
import matplotlib.pyplot as plt

### 이미지 크기 조정
image = cv2.imread('./data/image.jpg')
resize_image = cv2.resize(image, (200, 200))   # 크기를 640x480으로 조정

plt.imshow(resize_image)
plt.axis('off')  # 축 제거

cv2.imshow('Image', resize_image)
cv2.waitKey(0)
cv2.destroyAllWindows()  # 모든 창 닫기
```

```python
### 엣지 검출

# 이미지 불러오기
image = cv2.imread('./data/image.jpg')

# 흑백 변환
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 엣지 검출
# 100: 임계값1, 100: 임계값2 엣지 정도?인듯
# 임계값1은 낮을 수록, 임계값2는 높을 수록 엣지 검출 강함
edge_image = cv2.Canny(image, 50,300)

# 엣지 검출된 이미지 cv2로 출력
cv2.imshow('Edge Image', edge_image)
cv2.waitKey(0)
cv2.destroyAllWindows()  # 모든 창 닫기
```

```python
### 도형 그리기

# 이미지 불러오기
image = cv2.imread('./data/image.jpg')

# 이미지가 제대로 읽혔는지 확인
if image is None:
    print("이미지를 읽을 수 없습니다. 경로를 확인하세요")
else:
    # 사각형 그리기
    start = (50, 50)    # 시작 좌표 (x, y)
    end = (200, 200)    # 끝 좌표 (x, y)
    color = (255, 0, 0) # BGR
    thinckness = 2      # 선 두께
    cv2.rectangle(image, start, end, color, thinckness)

    # 원 그리기
    center_cordinates = (150, 150) # 원의 중심 좌표(x, y)
    radius = 50  # 원의 반지름
    color = (0, 255, 0)  # BGR
    thinckness = 2  # 선 두께
    cv2.circle(image, center_cordinates, radius, color, thinckness)

    # 선 그리기
    start_point = (100, 100)  # 시작 좌표 (x, y)
    end_point = (400, 400)  # 끝 좌표 (x, y)
    thinkcness = 5  # 선 두께
    cv2.line(image, start_point, end_point, color, thinkcness)

    # 텍스트 그리기
    text = "Hello"
    position = (200, 150)
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 1
    font_color = (0, 0, 255)
    thickness = 2
    cv2.putText(image, text, position, font, font_scale, font_color, thickness)
    
    # 도형이 그려진 이미지 출력
    cv2.imshow('Image with Shapes', image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()  # 모든 창 닫기
```
