### 프로젝트 과제2

[과제 주제]
> Red, Blue, Green 삼 색을 Detection 하여라.

```python
# 라이브러리 불러오기
import cv2
import numpy as np
```

```python
# 웹캠 열기
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("웹캠을 열 수 없습니다.")
    exit()

while(True):
    ret, frame = cap.read()
    if not ret:
        print("프레임을 읽을 수 없습니다.")
        break

    # HSV 색상 공간으로 변환
    hsv_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    ## 색상 범위 설정
    # 파란색
    lower_blue = np.array([100, 100, 100])
    upper_blue = np.array([140, 255, 255])

    # 초록색
    lower_green = np.array([40, 100, 100])
    upper_green = np.array([80, 255, 255])

    # 빨간색
    lower_red1 = np.array([0, 100, 100])
    upper_red1 = np.array([10, 255, 255])
    lower_red2 = np.array([160, 100, 100])
    upper_red2 = np.array([179, 255, 255])

    ## 색깔 영역 마스크 생성
    # 파란색
    blue_mask = cv2.inRange(hsv_frame, lower_blue, upper_blue)

    # 초록색
    green_mask = cv2.inRange(hsv_frame, lower_green, upper_green)

    # 빨간색
    red_mask1 = cv2.inRange(hsv_frame, lower_red1, upper_red1)
    red_mask2 = cv2.inRange(hsv_frame, lower_red2, upper_red2)
    red_mask = cv2.bitwise_or(red_mask1, red_mask2)

    # 마스크 윤곽 적용
    # 파란색 마스크의 윤곽
    contours, _ = cv2.findContours(blue_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        if cv2.contourArea(cnt) > 500:
            x,y,w,h = cv2.boundingRect(cnt)
            cv2.rectangle(frame, (x,y), (x+w, y+h), (255, 0, 0), 2)
            cv2.putText(frame, "Blue", (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255,0,0), 2)

    # 초록색 마스크의 윤곽
    contours, _ = cv2.findContours(green_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        if cv2.contourArea(cnt) > 500:
            x,y,w,h = cv2.boundingRect(cnt)
            cv2.rectangle(frame, (x,y), (x+w,y+h), (0,255,0), 2)
            cv2.putText(frame, "Green", (x,y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,255,0), 2)

    # 빨간색 마스크의 윤곽
    contours, _ = cv2.findContours(red_mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    for cnt in contours:
        if cv2.contourArea(cnt) > 500:
            x,y,w,h = cv2.boundingRect(cnt)
            cv2.rectangle(frame, (x,y), (x+w, y+h), (0,0,255), 2)
            cv2.putText(frame, "Red", (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0,0,255), 2)

    # 파란색, 초록색, 빨간색 마스크 합성
    combine_mask = cv2.bitwise_or(cv2.bitwise_or(green_mask, red_mask), blue_mask)

    # 결과 출력
    cv2.imshow("Original Fram", frame)
    cv2.imshow("Mask Fram", combine_mask)

    # ESC 키 누르면 종료
    if cv2.waitKey(1) & 0xFF == 27:
        break

# 자원 해제
cap.release()
cv2.destroyAllWindows()
```
