```python
# 라이브러리 불러오기
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
### 이미지 불러오기
image = cv2.imread('./data/rear_garden.png')
print(f"높이: {image.shape[0]}")
print(f"넓이: {image.shape[1]}")
print(f"채널 수: {image.shape[2]}")
```

```python
### 이미지를 그레이스케일로 변환
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# 이미지 저장
cv2.imwrite('gray_image.jpg', gray_image)

# 변환된 이미지 출력
plt.imshow(gray_image, cmap='gray') # 흑백으로 표시
plt.axis('off')     # 축 제거
plt.show()
```

```python
### 이미지 자르기
crop_image = image[50:150, 200:450]

# 이미지 저장
cv2.imwrite('cropped_image.jpg', crop_image)

# 이미지 출력
crop_image = cv2.cvtColor(crop_image, cv2.COLOR_BGR2RGB)
plt.imshow(crop_image)
plt.axis('off')
plt.show()
```

```python
## 이미지 도형 그리기

# 이미지가 제대로 읽혔는지 확인
if image is None:
    print("이미지를 읽어올 수 없습니다. 경로를 다시 확인해주세요.")
else:
    # 사각형 그리기
    start_s = (50, 50)      # 사각형 시작 좌표(x, y)
    end_s = (150, 150)      # 사각형 끝 좌표(x, y)
    color = (255, 255, 0)   # 색상
    thinkcness = 2          # 선 두께
    cv2.rectangle(image, start_s, end_s, color, thinkcness)

    # 원 그리기
    center = (300, 300)     # 원의 중심 좌표(x, y)
    radius = 50             # 반지름
    color = (0, 255, 255)   # 색상
    thinkcness = 2          # 선 두께
    cv2.circle(image, center, radius, color, thinkcness)

    # 선 그리기
    start_l = (100, 100)    # 선 시작 좌표(x, y)
    end_l = (400, 400)      # 선 끝 좌표(x, y)
    color = (255, 0, 255)   # 색상 
    thinkcness = 2          # 선 두께
    cv2.line(image, start_l, end_l, color, thinkcness)

    # 이미지 저장
    cv2.imwrite("shapes_image.jpg",image)

    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)  # matplotlib 출력을 위해 RGB로 변환
    plt.imshow(image)
    plt.axis('off')
    plt.show()
```

```python
## 여러 이미지를 하나의 창에 표시
image_path = ['gray_image.jpg', 'cropped_image.jpg', 'shapes_image.jpg']
images = []

for path in image_path:
    img = cv2.imread(path)
    img = cv2.resize(img, (300, 200))
    images.append(img)

if images:
    concate_image = np.hstack(images)
    cv2.imshow('Concated Image', concate_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("이미지를 로드할 수 없습니다.")
```

```python
import cv2
import numpy as np

# 표시할 이미지 파일 경로 리스트
image_paths = ['gray_image.jpg', 'cropped_image.jpg', 'shapes_image.jpg']
image_titles = ["Gray Image", "Cropped Image", "Shapes Image"]

# 불러온 이미지 객체(Numpy 배열)를 저장할 리스트
loaded_images = []

# 모든 이미지를 불러와 리스트에 추가
for path in image_paths:
    img = cv2.imread(path)
    if img is not None:
        # hstack을 위해 모든 이미지의 높이를 200으로 통일 (너비는 300)
        resized_img = cv2.resize(img, (300, 200))
        loaded_images.append(resized_img)
    else:
        print(f"경고: 이미지를 불러올 수 없습니다: {path}")

# 성공적으로 불러온 이미지가 있는지 확인
if loaded_images:
    # 리스트에 있는 모든 이미지를 수평으로 연결
    concatenated_image = np.hstack(loaded_images)
    
    # 최종적으로 합쳐진 이미지를 하나의 창에 표시
    cv2.imshow('All Images in One Window', concatenated_image)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
else:
    print("이미지를 로드할 수 없습니다.")

```
